```python
var ("x")
f = (2^sin(2*x) + arctan(2*x))^2 - 8*sin(x)
show(f)
```

```python
plot_f = plot(f, xmin = 1, xmax = 5, ymax = 20, ymin = -5)
show(plot_f)
```

```python
def newton(f, a, b, EPS):
    x1 = a
    x2 = (a + b) / 2
    
    df = f.derivative()

    while abs(x2 - x1) >= EPS:
        x1 = x2
        x2 = x1 - f(x=x1)/df(x=x1)

    return x1
```

```python
EPS = 0.01
a = 1
b = 5

x_1 = newton(f, a, b, EPS)

p1 = plot(f, xmin = 1, xmax = 5, ymax = 20, ymin = -5)
p1 += point((x_1, f(x=x_1)), color="red", size=30, zorder=20)

show(p1)
```

```python
p2 = plot_f
trapeze_square = 0
trapezes = []
for i, x_0 in enumerate(np.arange(a, b, dx)):
    dy1 = f(x_0)
    if (x_0 + dx <= b):
        c = x_0 + dx
    else:
        c = b
    dy2 = f(c)
    if (x_0 + dx <= b):
        curr = dx * ((dy1 + dy2) / 2)
    else:
        curr = (x_0 + dx - b) * ((dy1 + dy2) / 2)
    trapeze_square += curr
    txt = text(r"$i={}, curr={}, result={}$".format(i, str(curr.n(digits=6)), str(trapeze_square.n(digits=6))), (3.5, 1.2), fontsize=12, color="black")
    
    trapezes.append(
        polygon2d([(x_0, 0), (c, 0), (c, dy2), (x_0, dy1)], fill=False, color="red"))
    
    p2 += plot(trapezes[-1])
    if (i == 0):
        show(p2 + txt, xmin = (2*e+1)*e^(-1) - 0.05, xmax=e^2+2 - 5)
    if (i == 10):
        show(p2 + txt, xmin = (2*e+1)*e^(-1) - 0.05, xmax=e^2+2 - 5)
show(p2 + txt, xmin = (2*e+1)*e^(-1) - 0.05, xmax=e^2+2 - 5)
```
